```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
from sklearn.metrics import mean_squared_error
from numpy.polynomial.polynomial import Polynomial

# generate the data set
x_train = np.linspace(0,10,100)
y_train = np.array([math.sin(data) for data in x_train])
x_test = np.linspace(0,10,50)
y_test = np.array([math.sin(data) for data in x_test])

# sample 15 points from the training set
x_s = np.random.choice(x_train, size=15, replace=False,)
y_s = np.array([math.sin(data) for data in x_s])

# fit
poly = lagrange(x_s, y_s)

# predict
f = Polynomial(poly.coef[::-1])
y_train_ = f(x_train)
y_test_ = f(x_test)

# plot the results
plt.figure(figsize = (10,5))

plt.subplot(221)
plt.scatter(x_train, y_train, label='training data')
plt.scatter(x_train, y_train_, label='Polynomial', c = 'r',marker = '+')
plt.ylim(-2,2)
plt.legend()

plt.subplot(222)
plt.scatter(x_test, y_test, label='test data')
plt.scatter(x_test, y_test_, label='Polynomial', c = 'r', marker = '+')
plt.ylim(-2,2)
plt.legend()

plt.show()

print("the mean square error of training set is: ", mean_squared_error(y_train, y_train_))
print("the mean square error of test set is: ", mean_squared_error(y_test, y_test_))
```
